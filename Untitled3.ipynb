# Numerisk Derivasjon

```python
from pylab import *
tid = []
fart = []
akselerasjon = []
fil = open("bevegelse.txt", "r")

for linje in fil:
    fata = linje.split(",")
    tid.append(float(fata[0]))
    fart.append(float(fata[1]))

fil.close()
plot(tid, fart)
show()
def f(x):
    return x**2
def NewKvo(x, h):
    fder = (f(x+h)-f(x))/h
    print(fder)

for i in range(0, len(tid)):
    akselerasjon.append((fart[i]-fart[i-1])/(tid[i]-tid[i-1]))
plot(tid, akselerasjon)
show()
```

Når man deriverer numerisk finner man avstanden mellom to nærliggente punkter. 
Her er den første grafen farten til en som er på joggetur. Den andre grafen er akselerasjonen hans. 
For og finne akselerasjonen må man ta to nærliggende punkter fra den orginale grafen, og trekke den ene fra den andre. Deretter legger man den nye verdien inn som et punkt i den nye grafen. 
Da finner man hvor mye farten endrer seg .

```python
bilde = imread("bilde.png")
imshow(bilde)
show()
dfdx = abs(bilde[1:-1,2:, :] - bilde[1:-1,1:-1,:])
imshow(3*dfdx)
show()
dfdy = abs(bilde[2:,1:-1,:]-bilde[1:-1,1:-1,:])
imshow(3*dfdy)
show()

der = sqrt(dfdx**2+dfdy**2)
imshow(3*der)
show() 
```
